## Linear Advection

```python
eqn = 'LinAdv'
LA_mat = scipy.io.loadmat('LinAdv_ConvgData.mat') 
LA_dts = np.array(LA_mat['dts']); LA_U_Err = np.array(LA_mat['U_Err']); LA_dU_Err = np.array(LA_mat['dU_Err'])
```

```python
#plotting
Marks = ['o','s','+','*','^','>']
lines = ["-","--",":"]
colors = ["black","red","blue","gray"]

LA_st = [5,4,6,5,5,3,5,3,3];
LA_en = [9,7,9,9,9,7,9,7,7];
Cof = [1e-1,9e-2,2e+0,5e-3,6e-2,2e+0,1e-4,5e-2,4e-2];
Sl = [1,2,3,2,3,4,2,4,5];

# ERK-(s,p,q) scheme
S=[3,4,5,4,6,7,7,8,9]; 
P=[3,3,3,4,4,4,5,5,5]; 
Q=[1,2,3,1,3,4,1,4,5]; 

for i in range(3):
    fig = plt.figure(figsize = figSize)
    # u
    plt.plot(LA_dts[3*i,:],LA_U_Err[3*i,:],color=colors[0], marker=Marks[0],linewidth=2, linestyle=lines[0],label = "(%d,%d,%d)-u"%(S[3*i],P[3*i],Q[3*i]))
    plt.plot(LA_dts[3*i+1,:],LA_U_Err[3*i+1,:],color=colors[1], marker=Marks[0],linewidth=2, linestyle=lines[0],label = "(%d,%d,%d)-u"%(S[3*i+1],P[3*i+1],Q[3*i+1]))
    plt.plot(LA_dts[3*i+2,:],LA_U_Err[3*i+2,:],color=colors[2], marker=Marks[0],linewidth=2, linestyle=lines[0],label = "(%d,%d,%d)-u"%(S[3*i+2],P[3*i+2],Q[3*i+2]))
    #u_x
    plt.plot(LA_dts[3*i,:],LA_dU_Err[3*i,:],color=colors[0], marker=Marks[1], linewidth=2,linestyle=lines[1],label = "(%d,%d,%d)-$u_x$"%(S[3*i],P[3*i],Q[3*i]))
    plt.plot(LA_dts[3*i+1,:],LA_dU_Err[3*i+1,:],color=colors[1], marker=Marks[1], linewidth=2,linestyle=lines[1],label = "(%d,%d,%d)-$u_x$"%(S[3*i+1],P[3*i+1],Q[3*i+1]))
    plt.plot(LA_dts[3*i+2,:],LA_dU_Err[3*i+2,:],color=colors[2], marker=Marks[1], linewidth=2,linestyle=lines[1],label = "(%d,%d,%d)-$u_x$"%(S[3*i+2],P[3*i+2],Q[3*i+2]))
 
    plt.plot(LA_dts[3*i,LA_st[3*i]:LA_en[3*i]],Cof[3*i]*LA_dts[3*i,LA_st[3*i]:LA_en[3*i]]**Sl[3*i],linewidth=2,linestyle=lines[2],color=colors[3],label = "Slope %d"%(Sl[3*i]))
    plt.plot(LA_dts[3*i+1,LA_st[3*i+1]:LA_en[3*i+1]],Cof[3*i+1]*LA_dts[3*i+1,LA_st[3*i+1]:LA_en[3*i+1]]**Sl[3*i+1],linewidth=2,linestyle=lines[1],color=colors[3],label = "Slope %d"%(Sl[3*i+1]))
    plt.plot(LA_dts[3*i+2,LA_st[3*i+2]:LA_en[3*i+2]],Cof[3*i+2]*LA_dts[3*i+2,LA_st[3*i+2]:LA_en[3*i+2]]**Sl[3*i+2],linewidth=2,linestyle=lines[0],color=colors[3],label = "Slope %d"%(Sl[3*i+2]))

    plt.xscale("log"); plt.yscale("log"); plt.xlabel('$\Delta t$'); plt.ylabel('Error')
    plt.legend(loc='upper center', bbox_to_anchor=(0.5,1.45),ncol=2,fontsize=fs)
    plt.savefig('./Figures/%s_ConvPlotOrder_%d.pdf'%(eqn,P[3*i+i]),format='pdf', bbox_inches="tight",transparent=True)
```
