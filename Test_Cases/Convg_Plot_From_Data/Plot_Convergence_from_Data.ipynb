## This code produces all the convergence plots presented in the manuscript using the data.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
import os, sys
```

```python
# If path doesnâ€™t exist we create a new path
import os
path = './Figures/'
if not os.path.exists(path):
   os.makedirs(path)
```

```python
font = {#'family' : 'normal',
'weight' : 'normal',
'size'   : 18}
plt.rc('font', **font) 

fs = 14; nl = 2; figSize = (4.5, 5)
```

## Linear Advection

```python
eqn = 'LinAdv'
LA_mat = scipy.io.loadmat('LinAdv_ConvgData.mat') 
LA_dts = np.array(LA_mat['dts'][0]); LA_U_Err = np.array(LA_mat['U_Err']); LA_dU_Err = np.array(LA_mat['dU_Err'])
```

```python
#plotting
Marks = ['o','s','+','*','^','>']
lines = ["-","--",":"]
colors = ["black","red","blue","gray"]

LA_st = [5,4,6,5,5,3,5,2,1];
LA_en = [9,7,9,9,9,6,9,5,5];
Cof = [1e-1,9e-2,2e+0,5e-3,6e-2,3e+0,1e-4,5e-2,5e-2];
Sl = [1,2,3,2,3,4,2,4,5];

# ERK-(s,p,q) scheme
S=[3,4,5,4,6,7,7,9,10]; 
P=[3,3,3,4,4,4,5,5,5]; 
Q=[1,2,3,1,3,4,1,4,5]; 

for i in range(3):
    fig = plt.figure(figsize = figSize)
    # u
    plt.plot(LA_dts,LA_U_Err[3*i,:],color=colors[0], marker=Marks[0],linewidth=2, linestyle=lines[0],label = "(%d,%d,%d)-u"%(S[3*i],P[3*i],Q[3*i]))
    plt.plot(LA_dts,LA_U_Err[3*i+1,:],color=colors[1], marker=Marks[0],linewidth=2, linestyle=lines[0],label = "(%d,%d,%d)-u"%(S[3*i+1],P[3*i+1],Q[3*i+1]))
    plt.plot(LA_dts,LA_U_Err[3*i+2,:],color=colors[2], marker=Marks[0],linewidth=2, linestyle=lines[0],label = "(%d,%d,%d)-u"%(S[3*i+2],P[3*i+2],Q[3*i+2]))
    #u_x
    plt.plot(LA_dts,LA_dU_Err[3*i,:],color=colors[0], marker=Marks[1], linewidth=2,linestyle=lines[1],label = "(%d,%d,%d)-$u_x$"%(S[3*i],P[3*i],Q[3*i]))
    plt.plot(LA_dts,LA_dU_Err[3*i+1,:],color=colors[1], marker=Marks[1], linewidth=2,linestyle=lines[1],label = "(%d,%d,%d)-$u_x$"%(S[3*i+1],P[3*i+1],Q[3*i+1]))
    plt.plot(LA_dts,LA_dU_Err[3*i+2,:],color=colors[2], marker=Marks[1], linewidth=2,linestyle=lines[1],label = "(%d,%d,%d)-$u_x$"%(S[3*i+2],P[3*i+2],Q[3*i+2]))
 
    plt.plot(LA_dts[LA_st[3*i]:LA_en[3*i]],Cof[3*i]*LA_dts[LA_st[3*i]:LA_en[3*i]]**Sl[3*i],linewidth=2,linestyle=lines[2],color=colors[3],label = "Slope %d"%(Sl[3*i]))
    plt.plot(LA_dts[LA_st[3*i+1]:LA_en[3*i+1]],Cof[3*i+1]*LA_dts[LA_st[3*i+1]:LA_en[3*i+1]]**Sl[3*i+1],linewidth=2,linestyle=lines[1],color=colors[3],label = "Slope %d"%(Sl[3*i+1]))
    plt.plot(LA_dts[LA_st[3*i+2]:LA_en[3*i+2]],Cof[3*i+2]*LA_dts[LA_st[3*i+2]:LA_en[3*i+2]]**Sl[3*i+2],linewidth=2,linestyle=lines[0],color=colors[3],label = "Slope %d"%(Sl[3*i+2]))

    plt.xscale("log"); plt.yscale("log"); plt.xlabel('$\Delta t$'); plt.ylabel('Error')
    plt.legend(loc='upper center', bbox_to_anchor=(0.5,1.45),ncol=2,fontsize=fs)
    plt.savefig('./Figures/%s_ConvPlotOrder_%d.pdf'%(eqn,P[3*i+i]),format='pdf', bbox_inches="tight",transparent=True)
```

```python
#plotting
Marks = ['o','s','+','*','^','>']
lines = ["-","--",":"]
colors = ["black","red","blue","gray"]

B_st = [6,6,6,6,6,6,3,3,3];
B_en = [9,9,9,9,9,9,9,10,7];
Cof = [8e-3,5e-3,5e-1,1e-3,4e-3,5e-3,8e-5,8e-5,9e-5];
Sl = [1,2,3,1,2,3,1,2,3];

# ERK-(s,p,q) scheme
S=[3,4,5,4,6,7,7,9,10]; 
P=[3,3,3,4,4,4,5,5,5]; 
Q=[1,2,3,1,3,4,1,4,5]; 


for i in range(3):
    fig = plt.figure(figsize = figSize)
    # u
    plt.plot(B_dts,B_U_Err[3*i,:],color=colors[0], marker=Marks[0],linewidth=2, linestyle=lines[0],label = "(%d,%d,%d)-u"%(S[3*i],P[3*i],Q[3*i]))
    plt.plot(B_dts,B_U_Err[3*i+1,:],color=colors[1], marker=Marks[0],linewidth=2, linestyle=lines[0],label = "(%d,%d,%d)-u"%(S[3*i+1],P[3*i+1],Q[3*i+1]))
    plt.plot(B_dts,B_U_Err[3*i+2,:],color=colors[2], marker=Marks[0],linewidth=2, linestyle=lines[0],label = "(%d,%d,%d)-u"%(S[3*i+2],P[3*i+2],Q[3*i+2]))
    #u_x
    plt.plot(B_dts,B_dU_Err[3*i,:],color=colors[0], marker=Marks[1], linewidth=2,linestyle=lines[1],label = "(%d,%d,%d)-$u_x$"%(S[3*i],P[3*i],Q[3*i]))
    plt.plot(B_dts,B_dU_Err[3*i+1,:],color=colors[1], marker=Marks[1], linewidth=2,linestyle=lines[1],label = "(%d,%d,%d)-$u_x$"%(S[3*i+1],P[3*i+1],Q[3*i+1]))
    plt.plot(B_dts,B_dU_Err[3*i+2,:],color=colors[2], marker=Marks[1],linewidth=2, linestyle=lines[1],label = "(%d,%d,%d)-$u_x$"%(S[3*i+2],P[3*i+2],Q[3*i+2]))
 
    plt.plot(B_dts[B_st[3*i]:B_en[3*i]],Cof[3*i]*B_dts[B_st[3*i]:B_en[3*i]]**Sl[3*i],linewidth=2,linestyle=lines[2],color=colors[3],label = "Slope %d"%(Sl[3*i]))
    plt.plot(B_dts[B_st[3*i+1]:B_en[3*i+1]],Cof[3*i+1]*B_dts[B_st[3*i+1]:B_en[3*i+1]]**Sl[3*i+1],linewidth=2,linestyle=lines[1],color=colors[3],label = "Slope %d"%(Sl[3*i+1]))
    plt.plot(B_dts[B_st[3*i+2]:B_en[3*i+2]],Cof[3*i+2]*B_dts[B_st[3*i+2]:B_en[3*i+2]]**Sl[3*i+2],linewidth=2,linestyle=lines[0],color=colors[3],label = "Slope %d"%(Sl[3*i+2]))

    plt.xscale("log"); plt.yscale("log"); plt.xlabel('$\Delta t$'); plt.ylabel('Error')
    plt.legend(loc='upper center', bbox_to_anchor=(0.5,1.45),ncol=nl,fontsize=fs)
    plt.savefig('./Figures/%s_ConvPlotOrder_%d.pdf'%(eqn,P[3*i+i]),format='pdf', bbox_inches="tight",transparent=True)
```

## SWE in 1D

```python
tfinal=0.5; cfl = 0.8 
```

```python
dts = np.load("./SWE_1D_ConvgData/SWE_1D_dts_CFL_%1.1f_T_%1.1f.npy"%(cfl,tfinal), allow_pickle=True)
Err_u = np.load("./SWE_1D_ConvgData/SWE_1D_Err_u_CFL_%1.1f_T_%1.1f.npy"%(cfl,tfinal), allow_pickle=True)
Err_hu = np.load("./SWE_1D_ConvgData/SWE_1D_Err_hu_CFL_%1.1f_T_%1.1f.npy"%(cfl,tfinal), allow_pickle=True)
```

```python
S = [3,5,4,7,7,10];P =[3,3,4,4,5,5]; Q =[1,3,1,4,1,5]; 
#plotting
tick_positions = np.linspace(min(dts), max(dts), 3)
custom_ticks = np.linspace(min(dts), max(dts), 3)
Marks = ['o','s','+','*','^','>']
lines = ["-","--",":"]
colors = ["black","blue","red","gray"]

Sl = [2,3,2,3,2,3]; Coeff = [5e-1,1e0,2e-2,5e-2,5e-4,8e-2]; st = 3; en = 7;

for i in range(3):
    fig = plt.figure(figsize = figSize)
    plt.plot(dts,Err_u[2*i,:],color=colors[0], marker=Marks[0], linestyle=lines[0],linewidth=2, label = "(%d,%d,%d)-h"%(S[2*i],P[2*i],Q[2*i]))
    plt.plot(dts,Err_hu[2*i,:],color=colors[0], marker=Marks[0], linestyle=lines[1],linewidth=2,label = "(%d,%d,%d)-hu"%(S[2*i],P[2*i],Q[2*i]))

    plt.plot(dts,Err_u[2*i+1,:],color=colors[1], marker=Marks[1], linestyle=lines[0],linewidth=2,label = "(%d,%d,%d)-h"%(S[2*i+1],P[2*i+1],Q[2*i+1]))
    plt.plot(dts,Err_hu[2*i+1,:],color=colors[1], marker=Marks[1], linestyle=lines[1],linewidth=2,label = "(%d,%d,%d)-hu"%(S[2*i+1],P[2*i+1],Q[2*i+1]))
    
    plt.plot(dts[st:en],Coeff[2*i]*dts[st:en]**Sl[2*i],linestyle=lines[1], color="gray",linewidth=2,label='slope %d'%(Sl[2*i]))
    plt.plot(dts[st:en], Coeff[2*i+1]*dts[st:en]**Sl[2*i+1],linestyle=lines[0], color="gray",linewidth=2,label='slope %d'%(Sl[2*i+1]))

    plt.xscale("log"); plt.yscale("log"); plt.xlabel('$\Delta t$'); plt.ylabel('Error')
    plt.legend(loc='upper center', bbox_to_anchor=(0.5,1.3),ncol=nl,fontsize=fs)
    plt.savefig('./Figures/SWE1D_Zero_SpErr_ErrCong_p%d_t%1.1f.pdf'%(P[2*i],tfinal),format='pdf', bbox_inches="tight",transparent=True)
```
